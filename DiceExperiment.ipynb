```python

import random as r
import matplotlib.pyplot as plt

#N=int(input("Enter the number of times you want to perform the experiment with the dice : "))
#n=int(input("Enter the number of times dice will be thrown in 1 experiment : "))

  
def plotForDice(N,n,numBar):
    meanReadings=[]
    values=[]
    frequency=[]
    minreading=1.0
    maxreading=6.0
  
    for i in range(N):
        mean=0.0
        for j in range(n):
            mean+=r.randint(1,6)
        if mean<minreading :
            minreading =mean
        if mean>maxreading :
            maxnreading =mean
            
        
        meanReadings.append(mean/n)
   
    width=(maxreading-minreading)/numBar
    for i in range(0,numBar):
        minn=minreading+i*width
        maxx=minreading+(i+1)*width
        f=0 
        for x in meanReadings:
            if minn<=x<maxx :
                f+=1
                meanReadings.remove(x)
        frequency.append(f*1.0/N)
        values.append((minn+maxx)/2)
    plt.bar(values,frequency ,width)    
    plt.show()
```

```python
plotForDice(1000,5,10)
```
