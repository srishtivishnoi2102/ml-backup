```python
import random as r
import matplotlib.pyplot as plt

# N = number of times you want to perform the experiment with the coin 
# n = Number of times coin will be tossed in 1 experiment
# HEAD represented by 1
# TAIL represented by 2
def COIN_TOSSING_RV(N,n):
    # RV is the relative frequency
    RV=[]
    Xvalues=[]
    for a in range(n+1):
        RV.append(0)    # RV now becomes the list of frequency for each  0<=x<=n, all initialised with 0 for now
        Xvalues.append(a)  # Xvalues is the list for X,  0<=X<=n
    for i in range(N):
        f=0
        for j in range(n):
            if r.randint(1,2)==1 :    # if we are getting 1, i.e., HEAD 
                f+=1              # then increament the frequency for it
        # now f will tell us how many times out of n, we get the HEAD
        RV[f]+=(1.0/N)
        #in the RV list, the item corresponding to X=f will be incremented by 1/N, relative frequency 

    plt.bar(Xvalues,RV)
    plt.xlabel("X")
    plt.ylabel("RV(X)")
    plt.show()
    
def fact(n):
    if n==0 or n==1:
        return 1
    else:
        return n*fact(n-1)
    
def nCr(n,r):
    return int(fact(n)/(fact(n-r) * fact(r)))

# n= number of times coin is tossed 
# p= 

def COIN_TOSSING_PROBABILITY(n,p):
    xvalues=[]
    prob=[]
    for i in range(n+1):
        xvalues.append(i)
        prob.append(nCr(n,i) * p**i * (1-p)**(n-i) )
       # nCr(n,i) * p**i * (1-p)**(n-i)
        
    plt.bar(xvalues,prob) 
    plt.xlabel("X")
    plt.ylabel("P(X)")
    plt.show()
    
    
COIN_TOSSING_RV(50000,5)
COIN_TOSSING_PROBABILITY(5,.5)
```
