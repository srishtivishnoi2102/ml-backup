```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random as r 
from IPython.display import clear_output

from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
```

```python
data=pd.read_csv("datasets/ronaldo_data.csv")
```

```python
data=data.drop(['Unnamed: 0','team_name','team_id','knockout_match','date_of_game','type_of_shot'],axis=1)
```

```python
data.columns
```

# encoding home/away column

```python
t=data['home/away']
data['home/away']=t[t.isna()]=0
```

```python
t=[]
for i in range(len(data['home/away'])):

    if data['home/away'].iloc[i] !=0:
        if "@" in data['home/away'].iloc[i]:
            t.append(0)
        else:
            t.append(1)
    else:
        t.append(0)
data['home/away']=t
```

# DEALING WITH Game Season VALUES

```python
plt.hist(data['game_season'])
```

```python
data=data.replace(list(set(data['game_season'])),[i for i in range(len(set(data['game_season'])))])
data=data.replace(list(set(data['area_of_shot'])),[i for i in range(len(set(data['area_of_shot'])))])
data=data.replace(list(set(data['shot_basics'])),[i for i in range(len(set(data['shot_basics'])))])
data=data.replace(list(set(data['range_of_shot'])),[i for i in range(len(set(data['range_of_shot'])))])
data=data.replace(list(set(data['type_of_combined_shot'])),[i for i in range(len(set(data['type_of_combined_shot'])))])
data=data.replace(list(set(data['home/away'])),[i for i in range(len(set(data['home/away'])))])
data=data.replace(list(set(data['type_of_shot'])),[i for i in range(len(set(data['type_of_shot'])))])
```

```python
data=data[data['is_goal'].isna()==False]       
```

```python
for col in data.columns:
    try:
        
        data[col]=data[col].fillna(data[col].median())
    except:
        pass
```
