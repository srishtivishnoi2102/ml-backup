```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as s
```

```python
def abplot(a,b):
# a is the number of times exp done
# b is the number of times dice is thrown in 1 exp
    dicethrow= np.random.randint(1,7,size=(a,b))
    mean = np.mean(dicethrow,axis=1)
    median = np.median(dicethrow,axis=1)
    mode = s.mode(dicethrow,axis=1)

#print(mean)
#print(median)
#print(mode.mode)
#print(mode[0].shape)

    freqofmeans= s.itemfreq(mean)
    freqofmedians= s.itemfreq(median)
    freqofmodes= s.itemfreq(mode)

    plt.bar(freqofmeans[:,0],freqofmeans[:,1])
    plt.show()
    plt.bar(freqofmedians[:,0],freqofmedians[:,1])
    plt.show()
    plt.bar(freqofmodes[:,0],freqofmodes[:,1])
    plt.show()
```

```python
abplot(100000,6)
```
