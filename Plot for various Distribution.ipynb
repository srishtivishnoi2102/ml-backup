```python
import random as r
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def plot_Distribution(random_list,num_of_bar):
    random_list.sort()
    width = (random_list[-1]-random_list[0])/num_of_bar
    x_values=[random_list[0]+(i*width) for i in range(num_of_bar)]
    rel_freq_list=[0]*num_of_bar
    for i in range(len(random_list)):
        rel_freq_list[int((random_list[i]-random_list[0])//width)-1]+=(1/len(random_list))
    plt.bar(x_values,rel_freq_list)
    plt.plot(x_values,rel_freq_list,'r')
    plt.show()
```

# plot for data from normal distribution (np.random.randn):


```python
random_data=[np.random.randn() for a in range(0,100)]
plot_Distribution(random_data,5000)
```

# plot for random data (np.random.randint())...Plot tells that every number between the range is equally likely to occur


```python
random_data1=[np.random.randint(1,1000) for a in range(0,100000)]
plot_Distribution(random_data1,25)
```

# Plotting Poisson Distribution

```python
def poisson_Prob(x,mean):
    return ((2.71828)**mean * (mean)**x)/math.factorial(x)
```

```python
l=[i for i in range(100)]
probs1=[poisson_Prob(x,mean=25) for x in l ]
probs2=[poisson_Prob(x,mean=50) for x in l ]
```

```python
plt.scatter(l,probs1)
plt.show()
plt.scatter(l,probs2)
plt.show()
```

# observation: behaviour of poisson distribution is governed by the value of mean. That's why, mean is the natural parameter in poisson distribution
