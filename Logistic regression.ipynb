```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# import random as r
# import scipy.stats as S
rawdata = pd.read_csv("datasets/data.csv")
```

```python
y=rawdata['diagnosis'].values
x=rawdata[['radius_mean','texture_mean']].values
y=y.reshape(len(y),1)
```

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder=LabelEncoder()
y=labelEncoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2)

ytrain=ytrain.reshape(len(ytrain),1)
ytest=ytest.reshape(len(ytest),1)
```

```python
def sigmoid(z):
    return 1/(1+ np.exp(-z))
```

```python
# def logistic_loss(y_class,y_h_function_value):
#     return -np.mean( y_class*np.log(y_h_function_value) + (1-y_class)*np.log(1- y_h_function_value) )


def logistic_loss(y,h):    
    return -np.mean( y*np.log(h) + (1-y)*np.log(1-h) )
```

```python
weights_old=np.random.randn(2,1)
weights_new=weights_old
bias_old= np.random.randn(1,1)
bias_new=bias_old
learning_rate=0.01
```

```python
# training procedure
m=len(ytrain)
for each in range(5000):
    Z_old=np.matmul(xtrain,weights_old)+bias_old
    h_value_old=sigmoid(Z_old)
#     print(h_value_old)
    loss=logistic_loss(ytrain,h_value_old)
    print("loss in {}th iteration is: {}".format(each,loss))
    dZ=h_value_old-ytrain

#     bias_new = bias_old - learning_rate *   
    
    dWeights= np.matmul(xtrain.T,dZ) 
    dBias=np.sum(dZ)
    
    weights_new=weights_old - learning_rate*dWeights
    bias_new= bias_old - learning_rate*dBias

    Z_new=np.matmul(xtrain,weights_new)+bias_new
    h_value_new=sigmoid(Z_new)
    
    if abs(logistic_loss(ytrain,h_value_old)-logistic_loss(ytrain,h_value_new))<0.01:
        break
    bias_old=bias_new
    weights_old=weights_new
    
```
